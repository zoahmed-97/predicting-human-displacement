```python
#check and drop missing values
#replace entries for which building damage is missing with np.nan 

def clean_missing(df):
    
    df.drop(columns=['t0_mean_HTI_mat_prop_damage_3km',
                     't0_mean_dist2_DTMsites_km_Dec16Apr17',
                     't0_mean_sumDTM_IDPpop_3km_May17Sep17',
                     't0_mean_sumDTM_IDPpop_3km_Dec16Apr17'
                    ], inplace=True)
    
    df['t0_contact_locs'].apply(lambda x: x if x!= -999.00 else 0.0)
    
    
    if 't1_mean_HTI_mat_prop_damage_3km' in df.columns:
        df.drop(columns=['t1_mean_HTI_mat_prop_damage_3km',
                     't1_mean_dist2_DTMsites_km_Dec16Apr17',
                     't1_mean_sumDTM_IDPpop_3km_May17Sep17',
                     't1_mean_sumDTM_IDPpop_3km_Dec16Apr17',
                    't1_sum_idps_3km'
                    ], inplace=True)
        
        df['t1_contact_locs'].apply(lambda x: x if x!= -999.00 else 0.0)

    elif 't2_mean_HTI_mat_prop_damage_3km' in df.columns:
        df.drop(columns=['t2_mean_HTI_mat_prop_damage_3km',
                     't2_mean_dist2_DTMsites_km_Dec16Apr17',
                     't2_mean_sumDTM_IDPpop_3km_May17Sep17',
                     't2_mean_sumDTM_IDPpop_3km_Dec16Apr17',
                    't2_sum_idps_3km'], inplace=True)
        
        df['t2_contact_locs'].apply(lambda x: x if x!= -999.00 else 0.0)
        
    elif 't3_mean_HTI_mat_prop_damage_3km' in df.columns:
        df.drop(columns=['t3_mean_HTI_mat_prop_damage_3km',
                     't3_mean_dist2_DTMsites_km_Dec16Apr17',
                     't3_mean_sumDTM_IDPpop_3km_May17Sep17',
                     't3_mean_sumDTM_IDPpop_3km_Dec16Apr17',
                     't3_sum_idps_3km'
                    ], inplace=True)
        
        df['t3_contact_locs'].apply(lambda x: x if x!= -999.00 else 0.0)    
    #drop nan values
    df.dropna(inplace=True)    
    #drop duplicates
    df.drop_duplicates(inplace=True)
    #keep only one copy of IDPs with exact same combination of home location variables (to enable merging)
    df.drop_duplicates(subset=['t0_median_calls', 't0_rog', 't0_entropy_step', 't0_contact_locs',
       't0_known_locs', 't0_mean_HTI_mat_gust_footprint_3km',
       't0_mean_HTI_mat_prop_damage_3km_extrapdist',
       't0_mean_HTI_mat_rain3to5oct_mm_3km', 't0_mean_dist2_DTMsites_km_Nov16',
       't0_mean_dist2_mainroads_km', 't0_mean_dist2_urbancentres_km',
       't0_mean_popdensity3km_2015_UTM18N', 't0_mean_sumDTM_IDPpop_3km_Nov16'], inplace=True)
    
    df.reset_index(inplace=True)
    df.drop(columns='index', inplace=True)
    return df
```

#### Discretize the data

```python
from sklearn.preprocessing import KBinsDiscretizer

def discretize_into4(df_column):
    """Discretizes a column containing a continuous variable to one with a discrete one, 
    with 4 discrete categories"""
    discretizer = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')
    Y= df_column.values
    Y= discretizer.fit_transform(Y.reshape(len(Y),1))
    print("the number of bins for {} are {}".format(df_column.name, discretizer.n_bins_[0]))
    bin_edges= np.around(list(discretizer.bin_edges_)[0], decimals=2)
    intervals=[]
    for a, b in zip(bin_edges, bin_edges[1:]):
        intervals.append(str(a) + '-' + str(b))
    print("the edges of the bins are:", '[%s]'%', '.join(map(str, intervals)))
    return pd.Series(Y.flatten())

def discretize_into3(df_column):
    """Discretizes a column containing a continuous variable to one with a discrete one, 
    with 4 discrete categories"""
    discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
    Y= df_column.values
    Y= discretizer.fit_transform(Y.reshape(len(Y),1))
    print("the number of bins for {} are {}".format(df_column.name, discretizer.n_bins_[0]))
    print("the edges of the bins are:", discretizer.bin_edges_)
    return pd.Series(Y.flatten())


def discretize_into4_uniform(df_column):
    """Discretizes a column containing a continuous variable to one with a discrete one, 
    with 4 discrete categories"""
    discretizer = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
    Y= df_column.values
    Y= discretizer.fit_transform(Y.reshape(len(Y),1))
    print("the number of bins for {} are {}".format(df_column.name, discretizer.n_bins_[0]))
    bin_edges= np.around(list(discretizer.bin_edges_)[0], decimals=2)
    intervals=[]
    for a, b in zip(bin_edges, bin_edges[1:]):
        intervals.append(str(a) + '-' + str(b))
    print("the edges of the bins are:", '[%s]'%', '.join(map(str, intervals)))
    return pd.Series(Y.flatten())

def map_to_words(x, categories):
    """Maps ordinally encoded categories to bins (low, medium, high)"""
    if categories == 4:
        if x==0.0:
            return 'zero-low'
        if x==1.0:
            return 'low-medium'
        if x==2.0:
            return 'medium-high'
        if x==3.0:
            return 'high'
    elif categories == 3:
        if x==0.0:
            return 'low'
        if x==1.0:
            return 'medium'
        if x==2.0:
            return 'high'

def discretize_wind(x):
    """Discretizes the wind intensity column in line with Flowminder report"""
    if x<=50:
        return 'low'
    elif (x>50 and x<=70):
        return 'medium'
    elif (x>70):
        return 'high'

def discretize_rain(x):
    """Discretizes the rind intensity column in line with Flowminder report"""
    if x<=400:
        return 'low'
    elif (x>400 and x<=550):
        return 'medium'
    elif (x>550):
        return 'high'

def discretize_prop_damaged_buildings(x):
    """Discretizes the proportion of damaged buildings column in line with Flowminder report"""
    if x=='unknown':
        return x
    elif x<=0.6:
        return 'low'
    elif x>0.6:
        return 'high'
    
def discretize_known_locations(x):
    """Discretizes the known locations proportion within 10km column in line with Flowminder report"""
    if x<=0.001:
        return 'zero'
    elif (x>0.001 and x<=0.5):
        return 'some'
    elif (x>0.5):
        return 'most'
    
def discretize_contacts_prop(x):
    """Discretizes the contacts proportion within 10km column in line with Flowminder report"""
    if x<=0.001:
        return 'zero'
    elif (x>0.001 and x<=0.5):
        return 'some'
    elif (x>0.5):
        return 'most'
    

def discretize_DTMsites_distance(x):
    """Discretizes the distance from displacement sites NOT in line with Flowminder report"""
    if x<=10.0:
        return 'near'
    elif (x>10.0 and x<=50.0):
        return 'far'
    elif (x>50.0):
        return 'no_sites'
```

```python
#discretize the IDP specific metrics
df_all['t0_rog'] = discretize_into4(df_all['t0_rog']).apply(lambda x: map_to_words(x, 4))
df_all['t0_entropy_step'] = discretize_into4(df_all['t0_entropy_step']).apply(lambda x: map_to_words(x, 4))

#discretize the contact proportion (in lines with Flowminder report)
df_all['t0_contact_locs'] = df_all['t0_contact_locs'].apply(lambda x: discretize_contacts_prop(x))
df_all['t1_contact_locs'] = df_all['t1_contact_locs'].apply(lambda x: discretize_contacts_prop(x))
df_all['t2_contact_locs'] = df_all['t2_contact_locs'].apply(lambda x: discretize_contacts_prop(x))
df_all['t3_contact_locs'] = df_all['t3_contact_locs'].apply(lambda x: discretize_contacts_prop(x))


#discretize the known location proportion (in lines with Flowminder report)
df_all['t0_known_locs'] = df_all['t0_known_locs'].apply(lambda x: discretize_known_locations(x))
df_all['t1_known_locs'] = df_all['t1_known_locs'].apply(lambda x: discretize_known_locations(x))
df_all['t2_known_locs'] = df_all['t2_known_locs'].apply(lambda x: discretize_known_locations(x))
df_all['t3_known_locs'] = df_all['t3_known_locs'].apply(lambda x: discretize_known_locations(x))


#discretize the population density (not in lines with Flowminder report)
df_all['t0_popdensity3km_2015'] = discretize_into4(df_all['t0_popdensity3km_2015']).apply(lambda x: map_to_words(x, 4))
df_all['t1_popdensity3km_2015'] = discretize_into4(df_all['t1_popdensity3km_2015']).apply(lambda x: map_to_words(x, 4))
df_all['t2_popdensity3km_2015'] = discretize_into4(df_all['t2_popdensity3km_2015']).apply(lambda x: map_to_words(x, 4))
df_all['t3_popdensity3km_2015'] = discretize_into4(df_all['t3_popdensity3km_2015']).apply(lambda x: map_to_words(x, 4))


#discretize the proportion of damaged buildings (in lines with Flowminder report)
df_all['t0_prop_damaged_buildings'] = df_all['t0_prop_damaged_buildings'].apply(lambda x: discretize_prop_damaged_buildings(x))
df_all['t1_prop_damaged_buildings'] = df_all['t1_prop_damaged_buildings'].apply(lambda x: discretize_prop_damaged_buildings(x))
df_all['t2_prop_damaged_buildings'] = df_all['t2_prop_damaged_buildings'].apply(lambda x: discretize_prop_damaged_buildings(x))
df_all['t3_prop_damaged_buildings'] = df_all['t3_prop_damaged_buildings'].apply(lambda x: discretize_prop_damaged_buildings(x))

#discretize the wind intensity (in lines with Flowminder report)
df_all['t0_wind_instensity'] = df_all['t0_wind_instensity'].apply(lambda x: discretize_wind(x))
df_all['t1_wind_instensity'] = df_all['t1_wind_instensity'].apply(lambda x: discretize_wind(x))
df_all['t2_wind_instensity'] = df_all['t2_wind_instensity'].apply(lambda x: discretize_wind(x))
df_all['t3_wind_instensity'] = df_all['t3_wind_instensity'].apply(lambda x: discretize_wind(x))


#discretize the rain intensity (in lines with Flowminder report)
df_all['t0_rain_intensity'] = df_all['t0_rain_intensity'].apply(lambda x: discretize_rain(x))
df_all['t1_rain_intensity'] = df_all['t1_rain_intensity'].apply(lambda x: discretize_rain(x))
df_all['t2_rain_intensity'] = df_all['t2_rain_intensity'].apply(lambda x: discretize_rain(x))
df_all['t3_rain_intensity'] = df_all['t3_rain_intensity'].apply(lambda x: discretize_rain(x))

#discretize distance from main roads (not included in Flowminder report)
df_all['t0_dist2_mainroads'] = discretize_into4(df_all['t0_dist2_mainroads']).apply(lambda x: map_to_words(x, 4))
df_all['t1_dist2_mainroads'] = discretize_into4(df_all['t1_dist2_mainroads']).apply(lambda x: map_to_words(x, 4))
df_all['t2_dist2_mainroads'] = discretize_into4(df_all['t2_dist2_mainroads']).apply(lambda x: map_to_words(x, 4))
df_all['t3_dist2_mainroads'] = discretize_into4(df_all['t3_dist2_mainroads']).apply(lambda x: map_to_words(x, 4))


#discretize distance from urban centres (not included in Flowminder report)
df_all['t0_dist2_urbancentres'] = discretize_into4(df_all['t0_dist2_urbancentres']).apply(lambda x: map_to_words(x, 4))
df_all['t1_dist2_urbancentres'] = discretize_into4(df_all['t1_dist2_urbancentres']).apply(lambda x: map_to_words(x, 4))
df_all['t2_dist2_urbancentres'] = discretize_into4(df_all['t2_dist2_urbancentres']).apply(lambda x: map_to_words(x, 4))
df_all['t3_dist2_urbancentres'] = discretize_into4(df_all['t3_dist2_urbancentres']).apply(lambda x: map_to_words(x, 4))


#discretize the distance from displacement sites (not in line with Flowminder report)
#note that in contrast with report, an additional category of 'far' is added here
df_all['t0_dist2_DTMsites'] = df_all['t0_dist2_DTMsites'].apply(lambda x: discretize_DTMsites_distance(x))
df_all['t1_dist2_DTMsites'] = df_all['t1_dist2_DTMsites'].apply(lambda x: discretize_DTMsites_distance(x))
df_all['t2_dist2_DTMsites'] = df_all['t2_dist2_DTMsites'].apply(lambda x: discretize_DTMsites_distance(x))
df_all['t3_dist2_DTMsites'] = df_all['t3_dist2_DTMsites'].apply(lambda x: discretize_DTMsites_distance(x))

#discretize the number of IDPs in displacement camps (not included within Flowminder report)
df_all['t0_DTM_IDPpop'] = discretize_into4_uniform(df_all['t0_DTM_IDPpop']).apply(lambda x: map_to_words(x, 4))
df_all['t1_DTM_IDPpop'] = discretize_into4_uniform(df_all['t1_DTM_IDPpop']).apply(lambda x: map_to_words(x, 4))
df_all['t2_DTM_IDPpop'] = discretize_into4_uniform(df_all['t2_DTM_IDPpop']).apply(lambda x: map_to_words(x, 4))
df_all['t3_DTM_IDPpop'] = discretize_into4_uniform(df_all['t3_DTM_IDPpop']).apply(lambda x: map_to_words(x, 4))
```

```python

#discretize the distance from home metric (not in line with Flowminder report)
df_all['t1_dist_to_hl'] = discretize_into4(df_all['t1_dist_to_hl']).apply(lambda x: map_to_words(x, 4))
df_all['t2_dist_to_hl'] = discretize_into4(df_all['t2_dist_to_hl']).apply(lambda x: map_to_words(x, 4))
df_all['t3_dist_to_hl'] = discretize_into4(df_all['t3_dist_to_hl']).apply(lambda x: map_to_words(x, 4))
```
