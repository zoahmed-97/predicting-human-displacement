```python
df1= pd.read_csv('/Users/zo/Desktop/SDS_Mobility/Flowminder/displacements_counts/Haiti_Earthquake/t01.csv')
df2= pd.read_csv('/Users/zo/Desktop/SDS_Mobility/Flowminder/displacements_counts/Haiti_Earthquake/t02.csv')
df3= pd.read_csv('/Users/zo/Desktop/SDS_Mobility/Flowminder/displacements_counts/Haiti_Earthquake/t03.csv')
```

```python
#check and drop missing values
#replace entries for which building damage is missing with np.nan 

def clean_missing(df):
    df['t0_prop_damaged_buildings_300m']= df['t0_prop_damaged_buildings_300m'].apply(lambda x: x if x != -999.0 else 'unknown')
    
    if 't1_prop_damaged_buildings_300m' in df.columns:
        df['t1_prop_damaged_buildings_300m']= df['t1_prop_damaged_buildings_300m'].apply(lambda x: x if x != -999.0 else 'unknown')

    elif 't2_prop_damaged_buildings_300m' in df.columns:
        df['t2_prop_damaged_buildings_300m']= df['t2_prop_damaged_buildings_300m'].apply(lambda x: x if x != -999.0 else 'unknown')
    
    elif 't3_prop_damaged_buildings_300m' in df.columns:
        df['t3_prop_damaged_buildings_300m']= df['t3_prop_damaged_buildings_300m'].apply(lambda x: x if x != -999.0 else 'unknown')

    #drop nan values
    df.dropna(inplace=True)    
    #drop duplicates
    df.drop_duplicates(inplace=True)
    df.drop_duplicates(subset=['t0_rog', 't0_entropy_step', 't0_known_locs','t0_prop_damaged_buildings_300m', 't0_EQ_MMI',
       't0_dist2_DTMsites_km_2010', 't0_popdensity3km_2010',
       't0_sumDTM_IDPpop_300m_2010'],inplace=True)
    
    df.reset_index(inplace=True)
    df.drop(columns='index', inplace=True)
    return df
```

```python
df1, df2, df3= clean_missing(df1), clean_missing(df2), clean_missing(df3)
```

#### Merge Data

```python
df_all = pd.merge(df1, df2, on=['t0_rog', 't0_entropy_step', 't0_known_locs','t0_prop_damaged_buildings_300m', 't0_EQ_MMI',
       't0_dist2_DTMsites_km_2010', 't0_popdensity3km_2010',
       't0_sumDTM_IDPpop_300m_2010'], how='inner')
```

```python
df_all= pd.merge(df_all, df3, on=['t0_rog', 't0_entropy_step', 't0_known_locs','t0_prop_damaged_buildings_300m', 't0_EQ_MMI',
       't0_dist2_DTMsites_km_2010', 't0_popdensity3km_2010',
       't0_sumDTM_IDPpop_300m_2010'], how='inner')
```

#### Discretize the data

```python
from sklearn.preprocessing import KBinsDiscretizer

def discretize_into3(df_column):
    """Discretizes a column containing a continuous variable to one with a discrete one, 
    with 3 discrete categories"""
    discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
    Y= df_column.values
    Y= discretizer.fit_transform(Y.reshape(len(Y),1))
    print("the number of bins for {} are {}".format(df_column.name, discretizer.n_bins_[0]))
    bin_edges= np.around(list(discretizer.bin_edges_)[0], decimals=2)
    print("the edges of the bins are:", bin_edges)
    return pd.Series(Y.flatten())

def map_to_words(x):
    """Maps ordinally encoded categories to words (low, medium, high)"""
    if x==0.0:
        return 'low'
    if x==1.0:
        return 'medium'
    if x==2.0:
        return 'high'
    
    
def discretize_EQ(x):
    """Discretizes the EQ intensity column in line with Flowminder report"""
    if x<=5:
        return 'low'
    elif (x>5 and x<=7):
        return 'medium'
    elif (x>7):
        return 'high'

def discretize_prop_damaged_buildings(x):
    """Discretizes the proportion of damaged buildings column in line with Flowminder report"""
    if x=='unknown':
        return x
    elif x<=0.2:
        return 'low'
    elif x>0.2:
        return 'high'
    
def discretize_known_locations(x):
    """Discretizes the known locations proportion within 10km column in line with Flowminder report"""
    if x<=0.001:
        return 'zero'
    elif (x>0.001 and x<=0.5):
        return 'some'
    elif (x>0.5):
        return 'most'
    

def discretize_DTMsites_distance(x):
    """Discretizes the distance from displacement sites NOT in line with Flowminder report"""
    if x<=10.0:
        return 'near'
    elif (x>10.0 and x<=50.0):
        return 'far'
    elif (x>50.0):
        return 'no_sites'
```

```python
#discretize the IDP specific metrics
df_all['t0_rog'] = discretize_into3(df_all['t0_rog']).apply(lambda x: map_to_words(x))
print("bin edges for t0_entropy_step")
df_all['t0_entropy_step'] = discretize_into3(df_all['t0_entropy_step']).apply(lambda x: map_to_words(x))


#discretize the population density (not in lines with Flowminder report)
df_all['t0_popdensity3km_2010'] = discretize_into3(df_all['t0_popdensity3km_2010']).apply(lambda x: map_to_words(x))
df_all['t1_popdensity3km_2010'] = discretize_into3(df_all['t1_popdensity3km_2010']).apply(lambda x: map_to_words(x))
df_all['t2_popdensity3km_2010'] = discretize_into3(df_all['t2_popdensity3km_2010']).apply(lambda x: map_to_words(x))
df_all['t3_popdensity3km_2010'] = discretize_into3(df_all['t3_popdensity3km_2010']).apply(lambda x: map_to_words(x))


#discretize the proportion of damaged buildings (in lines with Flowminder report)
df_all['t0_prop_damaged_buildings_300m'] = df_all['t0_prop_damaged_buildings_300m'].apply(lambda x: discretize_prop_damaged_buildings(x))
df_all['t1_prop_damaged_buildings_300m'] = df_all['t1_prop_damaged_buildings_300m'].apply(lambda x: discretize_prop_damaged_buildings(x))
df_all['t2_prop_damaged_buildings_300m'] = df_all['t2_prop_damaged_buildings_300m'].apply(lambda x: discretize_prop_damaged_buildings(x))
df_all['t3_prop_damaged_buildings_300m'] = df_all['t3_prop_damaged_buildings_300m'].apply(lambda x: discretize_prop_damaged_buildings(x))

#discretize the EQ intensity (in lines with Flowminder report)
df_all['t0_EQ_MMI'] = df_all['t0_EQ_MMI'].apply(lambda x: discretize_EQ(x))
df_all['t1_EQ_MMI'] = df_all['t1_EQ_MMI'].apply(lambda x: discretize_EQ(x))
df_all['t2_EQ_MMI'] = df_all['t2_EQ_MMI'].apply(lambda x: discretize_EQ(x))
df_all['t3_EQ_MMI'] = df_all['t3_EQ_MMI'].apply(lambda x: discretize_EQ(x))



#discretize the known location proportion (in lines with Flowminder report)
df_all['t0_known_locs'] = df_all['t0_known_locs'].apply(lambda x: discretize_known_locations(x))
df_all['t1_known_locs'] = df_all['t1_known_locs'].apply(lambda x: discretize_known_locations(x))
df_all['t2_known_locs'] = df_all['t2_known_locs'].apply(lambda x: discretize_known_locations(x))
df_all['t3_known_locs'] = df_all['t3_known_locs'].apply(lambda x: discretize_known_locations(x))


#discretize the distance from displacement sites (not in line with Flowminder report)
#note that in contrast with report, an additional category of 'far' is added here
df_all['t0_dist2_DTMsites_km_2010'] = df_all['t0_dist2_DTMsites_km_2010'].apply(lambda x: discretize_DTMsites_distance(x))
df_all['t1_dist2_DTMsites_km_2010'] = df_all['t1_dist2_DTMsites_km_2010'].apply(lambda x: discretize_DTMsites_distance(x))
df_all['t2_dist2_DTMsites_km_2010'] = df_all['t2_dist2_DTMsites_km_2010'].apply(lambda x: discretize_DTMsites_distance(x))
df_all['t3_dist2_DTMsites_km_2010'] = df_all['t3_dist2_DTMsites_km_2010'].apply(lambda x: discretize_DTMsites_distance(x))

#discretize the number of IDPs in displacement camps (not included within Flowminder report)
df_all['t0_sumDTM_IDPpop_300m_2010'] = discretize_into3(df_all['t0_sumDTM_IDPpop_300m_2010']).apply(lambda x: map_to_words(x))
df_all['t1_sumDTM_IDPpop_300m_2010'] = discretize_into3(df_all['t1_sumDTM_IDPpop_300m_2010']).apply(lambda x: map_to_words(x))
df_all['t2_sumDTM_IDPpop_300m_2010'] = discretize_into3(df_all['t2_sumDTM_IDPpop_300m_2010']).apply(lambda x: map_to_words(x))
df_all['t3_sumDTM_IDPpop_300m_2010'] = discretize_into3(df_all['t3_sumDTM_IDPpop_300m_2010']).apply(lambda x: map_to_words(x))

#discretize the distance from home metric (not in line with Flowminder report)
df_all['t1_dist_to_hl'] = discretize_into3(df_all['t1_dist_to_hl']).apply(lambda x: map_to_words(x))
df_all['t2_dist_to_hl'] = discretize_into3(df_all['t2_dist_to_hl']).apply(lambda x: map_to_words(x))
df_all['t3_dist_to_hl'] = discretize_into3(df_all['t3_dist_to_hl']).apply(lambda x: map_to_words(x))
```
